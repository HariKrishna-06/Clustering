# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling = MinMaxScaler()
```

```python
norm_data = scaling.fit_transform(crime)
```

```python
norm_data
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])
```

```python
df_norm
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# create clusters
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(df_norm)
```

```python
y_hc
```

```python
crime['h_clusterid']=hc.labels_
```

# KMeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
data['Clust']=x
```

```python
data
```

```python
data.iloc[:,1:5].groupby(data.Clust).mean()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data.info()
```

```python
df=data.iloc[:,1:5]
```

```python
df
```

```python
df.values
```

```python
std_scaler = StandardScaler()
```

```python
std_df = std_scaler.fit_transform(df.values)
```

```python
std_df
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(std_df)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([data,cl],axis=1)
```
